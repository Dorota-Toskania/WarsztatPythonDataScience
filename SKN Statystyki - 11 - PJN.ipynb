## Regular Expression Syntax (Part 1)

##### Basic expressions

- `.` - matches _**ANY**_ single character except a newline
- `^` - matches the _**BEGINNING**_ of a string
- `$` - matches the _**END**_ of a string
- `[]` - matches the _**SET**_ of characters inside 
- `[^]` - does not match the _**SET**_ of characters inside 

```python
result = re.search('^a', 'abc')
print(result)
```

```python
result = re.search('[aeiou]', 'abc')
print(result)
```

```python
result = re.search('[^a]', 'abc')
print(result)
```

---
##### Repetitions

 
- `?` - matches _**0 or 1**_ repetitions of the previous expression
- `*` - matches _**0 or more**_ repetitions of the previous expression (*Kleene's star*)
- `+` - matches _**1 or more**_ repetitions of the previous expression (*Kleene's plus*)
- `{N}` - matches _**EXACTLY N**_ repetitions of the previous expression 
- `{N,}` - matches _**AT LEAST N**_ repetitions of the previous expression 
- `{N,M}` - matches _**BETWEEN N and M**_ repetitions of the previous expression 



```python
result = re.search('a*', 'aaaaaaaaabc')
print(result)
```

---
##### Character CLASSES

 
- `\d` - matches a single character that is a _**digit**_ (*\[0-9\]*)
- `\D` - matches a single character that is a _**non digit character**_ (*\[^0-9\]*)
- `\w` - matches a single character that is a _**word character**_ (alphanumeric character plus underscore *\[a-zA-Z0-9_\]*)
- `\W` - matches a single character that is a _**not a word character**_ 
- `\s` - matches a single character that is a _**whitespace**_ (ie. space, tabs and line breaks *\[ \t\n\r\f\v\]*)
- `\S` - matches a single character that is a _**not a whitespace**_ (*\[^ \t\n\r\f\v\]*)
- `\b` - matches the empty string at word end/beginning

```python
import gzip
import sys
import re

f = gzip.open('wyklad11/odm.txt.gz', 'rt', encoding='utf-8')
dictionary = {}

for x in f:
    t = x.strip().split(',')
    tt = [ x.strip().lower() for x in t]
    for w in tt[1:]: 
        dictionary[w]=tt[0]
```

```python
def lematize(w):
    return dictionary.get(w,w)
```

## Statystyki

```python
import re

splitter = re.compile(r'[^ąąćęńłóóśśżżź\w]+')

def preprocessing(opis):
    tokenized = splitter.split(opis)
    l = list(tokenized)
    l = [ x.lower() for x in l ]
    l = [ lematize(x) for x in l ]
    return l
```

```python
n=4
for i in dum_df.iterrows():
    print(i[1][1])
    print('*'*25)
    print()
    n-=1
    if n==0: break
```

```python
corpus=[]
n=4
for i in dum_df.iterrows():
    opis = i[1][1]
    l = preprocessing(opis)
    corpus.append(l)
    n-=1
    if n==0: break

print(corpus)
```

```python
corpus=[]
for i in dum_df.iterrows():
    l = preprocessing(i[1][1])
    corpus.append(l)

print(f"Mamy {len(corpus)} tekstów")
```

```python
all_words = []
for t in corpus:
    all_words += t
 
print(f"Mamy {len(all_words)} wyrazów")
all_words[:15]
```

```python
counter = {}

for w in all_words:
    counter[w] = counter.get(w,0)+1

print(f"Mamy {len(counter.keys())} RÓŻNYCH wyrazów")
```

```python
counted_words= [ (word,cnt) for word,cnt in counter.items() ]
counted_words[:4]
```

```python
from operator import itemgetter

counted_words.sort(key=itemgetter(1), reverse=True)
counted_words[:10]
```

## Metryka TF-IDF

ile razy występuję wyraz *i* w tekście *j*

$${n}_{ij}$$ 


 ### Term Frequency (TF)
 
 $${tf}_{ij} = \frac{{n}_{ij}}{\sum{k}{{n}_{ik}}}$$
 
 W tekście *j* sprawdzamy ile proporcjonalnie do całości występuje wyraz *i*

### Inverted Document Frequency (IDF)

 $$idf_i = log \frac{|D|}{ \{ d: n_i \in d \}}$$
 
 licznik - liczba dokumentów
 
 mianownik - liczba dokumentów w którym wystapił wyraz *i*-ty
